```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Ranking_ENEM_2024_Completo.xlsx")

plt.figure(figsize=(10,6))

df.boxplot(column="Média",
           by="Dependência Administrativa")

plt.title("Distribuição da Média Geral por Dependência Administrativa")
plt.suptitle("")
plt.xlabel("Dependência Administrativa")
plt.ylabel("Média Geral")

plt.tight_layout()
plt.savefig("boxplot_dependencia.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Ranking_ENEM_2024_Completo.xlsx")

media_estado = (
    df.groupby("Estado")["Média"]
      .mean()
      .sort_values(ascending=False)
)

plt.figure(figsize=(12,8))

media_estado.plot(kind="bar")

plt.title("Média Geral por Estado")
plt.xlabel("Estado")
plt.ylabel("Média Geral")

plt.tight_layout()
plt.savefig("media_estado.png")
plt.show()
```
